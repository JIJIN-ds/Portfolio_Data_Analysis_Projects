```python
# non linear model

c = np.array([22,33,41,21,33,65,21])
d = np.array([11,44,31,45,65,66,27])
```

```python
plt.title("Non linear model")
plt.xlabel("c")
plt.ylabel("d")
plt.plot(c,d)
plt.show()
```

```python
# to get the same data in doted lines

plt.title("Non linear model")
plt.xlabel("c")
plt.ylabel("d")
plt.plot(c,d,'--g')
plt.show()
```

```python
x = [5,10,15,10,8]
y = [10,15,20,21,60]
plt.bar(x,y)
plt.title("Bar Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# other parameters in bar
x=[5,6,7,8,9,10]
y=[12,10,14,8,16,6]
plt.bar(x,y,color='red',align='edge',edgecolor='blue',linewidth='0.8')
plt.title("Colorful bar graph")
plt.xlabel("no of people in group")
plt.ylabel("Range of work")
plt.show()
```

# Seaborn Visualization

```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x = ['sun','mon','tue','wed','thu','fri','sat']
y = [5,6,7,8.9,12,5.6,4.8]
```

```python
#Strip plot

a = sns.stripplot(x,y)
a.set(xlabel="Days",ylabel="Amount Spend")
plt.title("Days Vs Amount Spend")
plt.show()
```

```python
# upto thursday the amount spend increasing and then dropping all of a sudden on friday
```

```python
sns.violinplot?
```

```python
# violin plot

ds = pd.read_csv("emp.csv")
print(ds.head(5))

#violin plot
sns.violinplot(x="Salary",data=ds)

plt.show()
```

```python
sns.violinplot(ds["Salary"])
```

```python
a= sns.violinplot(x="Age",data=ds)
plt.show()
```

```python
ds_iris = pd.read_csv("iris.csv")
```

```python
ds_iris
```

```python
print(ds_iris.head(5))
sns.set(style = "whitegrid")
ax=sns.stripplot(x="class",y="sepal length",data=ds_iris)
ax.set(xlabel="Flower class")
plt.title("Graph representing sepal length belonging to class")
plt.show()
```

```python
print(ds_iris.loc[:,["sepal width","class"]])
sns.set(style="whitegrid")
ax= sns.stripplot(x="class",y="sepal width",data=ds_iris)
ax.set(xlabel="Iris class")
plt.title("Sepal width belonging to class")
plt.show()
```

```python
#swarmplot   -- more like stripplot but ordered manner
sns.swarmplot(x="class",y="sepal width",data=ds_iris)
```

```python
#changing orientation

sns.swarmplot(x="petal length",y="class",data=ds_iris)
```
