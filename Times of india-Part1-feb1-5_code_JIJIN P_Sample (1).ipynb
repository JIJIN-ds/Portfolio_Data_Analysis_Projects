```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://timesofindia.indiatimes.com/archive.cms"
driver.get(url)

time.sleep(2)
```

```python
#finding feb 2018 button
feb2018_button = driver.find_element_by_xpath('//a[@class="normtxt"][36]')
```

```python
#clicking on the button
feb2018_button.click()
```

```python
#creating an empty list to store each days url link
url=[]
```

```python
#since we have different tag name for each row of days . we extend aselecting multiple elements to get all url
url=driver.find_elements_by_xpath('//table[@id="calender"]/tbody/tr[2]/td/a')
```

```python
url.extend(driver.find_elements_by_xpath('//table[@id="calender"]/tbody/tr[3]/td/a'))
```

```python
url.extend(driver.find_elements_by_xpath('//table[@id="calender"]/tbody/tr[4]/td/a'))
```

```python
url.extend(driver.find_elements_by_xpath('//table[@id="calender"]/tbody/tr[5]/td/a'))
```

```python
url.extend(driver.find_elements_by_xpath('//table[@id="calender"]/tbody/tr[6]/td/a'))
```

```python
#we find length of url is 32 in which first 4 url is of empty space. we can remove it.
len(url)
```

```python
#concluding final urls
url=url[4:]
```

```python
len(url)
```

```python
#extracting each news link present in every url
news_link=[]

for i in url:
    news_link.append(i.get_attribute("href"))
    
```

```python
#we got newslink for all days. I tried doing all at once but data was lost due to power failure and internet issue. So i divided news_link into 6 parts from feb1-5 to feb 26-28 . This helped in extracting data
news_link=news_link[0:5]
```

```python
news link for feb1 - feb 5
news_link
```

```python
# All the news articles in the news link is extracted by
final=[]
for page in news_link:
    driver.get(page)
    time.sleep(10)
    
    news=driver.find_elements_by_xpath('//span[@style="font-family:arial ;font-size:12;color: #006699"]/a')
    
    for i in news:
        final.append(i.get_attribute("href"))
```

```python
#There are 2500 news article found in each of the 5 days
len(final)
```

```python
#link for each article present
final
```

```python
#creating empty list and storing data from each news article by this code

Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
fina=final

num=0
for f in fina:
    num=num+1
    print(num,"executing ...")
    driver.get(f)
    time.sleep(10)
    
    try:
        try:
            h=driver.find_element_by_xpath("//div[@class='as_heading']/h1/arttitle")
            Headline.append(h.text.replace("\n",''))
        
        except:
            h=driver.find_element_by_xpath("//h1[@class='_1Y-96'][1]")
            Headline.append(h.text.replace("\n",''))
            
    
    except NoSuchElementException :
        Headline.append("-")
        
        
    try:
        try:
            d=driver.find_element_by_xpath("//div[@data-plugin='dateformat']")
            Date.append(d.text.replace("Created: ",''))
        
        except:
            d=driver.find_element_by_xpath("//div[@class='yYIu- byline'][1]/span")
            Date.append(d.text.replace("Updated: ",''))
            
    
    except NoSuchElementException :
        Date.append("-")
        
        
    try:
        try:
            a=driver.find_element_by_xpath("//div[@itemprop='author']/a")
            Author.append(a.text)
        
        except:
            a=driver.find_element_by_xpath("//div[@class='yYIu- byline'][1]/a")
            Author.append(a.text)
            
    
    except NoSuchElementException :
        Author.append("-")
        
    try:
        try:
            v=driver.find_element_by_xpath("//div[@class='wrapper']/ol/li[2]/a/span")
            Vertical.append(v.text.replace("\n",''))
        
        except:
            v=driver.find_element_by_xpath("//div[@class='_1nPcO']/ul/li[2]/a/span")
            Vertical.append(v.text.replace("\n",''))
            
    
    except NoSuchElementException :
        Vertical.append("-")
    
    try:
        try:
            des=driver.find_element_by_xpath("//div[@class='Normal'][1]")
            Description.append(des.text.replace("\n",''))
        
        except:
            des=driver.find_element_by_xpath("//div[@class='_3YYSt clearfix  '][1]")
            Description.append(des.text.replace("\n",''))
            
    
    except NoSuchElementException :
        Description.append("-")
        
    
```
